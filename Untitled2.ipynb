```python
import numpy as np

def step(z):
  return 1 if  z >= 0 else 0

def neurona (x1,x2,w1,w2,b):
  z = x1*w1+x2*w2+b
  y = step(z)
  return z,y

# valores base

w1,w2,b = 1,1, -2.0 #intenta parecerse a una compuerta AND
casos = [(0,0), (0,1),(1,0),(1,1)]
for x1,x2 in casos:
  z,y = neurona(x1,x2,w1,w2,b)
  print(f"Entrada = ({x1},{x2}) z={z:.1f} salida={y}")


```

1. Con que valor de b la neurona se comporta mas parecido a AND?  
con -2.0

2. que efecto tiene bajar o subir el bias en la salida?  
entre mas alto sea el bias(mas cerca a 0) mas salidas hay en 1, entre mas bajo sea hay mas valores en 0
